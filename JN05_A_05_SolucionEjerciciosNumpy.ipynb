#### Importar NumPy como np

```python
import numpy as np
```

#### Crear un arreglo de 10 ceros 

```python
np.zeros(10)
```

#### Crear un arreglo de 10 unos

```python
np.ones(10)
```

#### Crear un arreglo de 10 cincos

```python
np.ones(10) * 5
```

#### Crear un arreglo con los enteros del 10 al 50

```python
np.arange(10,51)
```

#### Crear un arreglo con todos los enteros pares del 10 al 50

```python
np.arange(10,51,2)
```

#### Crear una matriz de 3x3 con valores que vayan del 0 al 8

```python
np.arange(9).reshape(3,3)
```

#### Crear una matriz identidad de 3x3

```python
np.eye(3)
```

#### Utilice NumPy para generar un número al azar entre 0 y 1

```python
np.random.rand(1)
```

#### Utilice NumPy para generar un arreglo de 25 números al azar tomados de una distribución estándard normal

```python
np.random.randn(25)
```

#### Crear la siguiente matriz:

```python
np.arange(1,101).reshape(10,10) / 100
```

#### Crear un arreglo de 20 puntos espaciados linealmente entre 0 y 1:

```python
np.linspace(0,1,20)
```
