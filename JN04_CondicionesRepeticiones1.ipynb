### Expresiones Booleanas
> Una expresión booleana es el conjunto de valores y operadores que dan, como resultado, **Verdadero** 👍🏼 (*True*) o **Falso** 👎🏼 (*False*).

A continuación se muestran algunas expresiones booleanas simples:

```python
print(2 < 5)
print(3 > 7)
print(5 <= 6)
print(4 >= 8)
```

```python
x = 11
print(x > 10)
print(2 * x < x)
```

Los resultados de estas expresiones son las *palabras reservadas* **True** o **False** (*¡sin comillas!*). Éstos son los únicos dos valores booleanos posibles. En Python, el nombre **booleano** se abrevia con la palabra **bool**. 

> True y False son del tipo de dato **booleano** o **bool**, no son un texto o cadena.


```python
print(type(True))
print(type(False))
```

Los operadores de comparación e igualdad que se pueden utilizar en Python son los siguientes:

```python
# inicialización de variables
x = 0
y = 1

# mostrar las operaciones de comparación e igualdad
print(x == y)  #x es igual a y
print(x != y)  #x es diferente de y
print(x > y)   #x es mayor que y
print(x < y)   #x es menor que y
print(x >= y)  #x es mayor o igual que y
print(x <= y)  #x es menor o igual que y
```

**Advertencia:** ¡Es un error común usar solamente un signo igual cuando se quiere evaluar la igualdad de dos valores! Recordemos que un signo de igual **=** se utiliza para asignación, es decir, guardar un valor en una variable. Para comparar igualdad debemos usar dos signos de igual juntos **==**. 🤓

### Ciclo while

>Se usa la instrucción **while** cuando se quiere repetir una serie de instrucciones mientras una expresión booleana de como resultado True.

**¡Importante!** Aquí las expresiones booleanas se evalúan una o más veces. ❗️

![Diagrama](https://www.tutorialspoint.com/python3/images/python_while_loop.jpg)

En lenguaje común es como decir los siguiente:

*Mientras el café esté demasiado caliente entonces, agregue un pequeño trozo de hielo*

1. ¿Está demasiado caliente el café? 
2. 😣(**True**) entonces agrega un trozo de hielo. 
3. Vuelve a probar el café, ¿todavía está demasiado caliente? 
4. 😣(**True**) entonces agrega otro trozo hielo. 
5. Vuelve a probar el café, ¿todavía está demasiado caliente? 
6. 😌(**False**) ya se lo puede tomar. 

Se **repite** el proceso de agregar trocitos de hielo mientras el café esté demasiado caliente, o en otras palabras, hasta que el café ya no esté demasiado caliente. 

Python tiene una sintaxis similar para esto:

>**while** exp_booleana:  
&nbsp;&nbsp;&nbsp;&nbsp;instrucciones con sangría<br/>
siguiente instrucción del programa

Cada vez que se llega al final de las instrucciones con sangría dentro del **while**, el programa vuelve al inicio del ciclo para probar si la expresión booleana sigue siendo verdadera. Únicamente cuando la expresión booleana es finalmente falsa, el flujo del programa salta, ignorando las instrucciones con sangría del ciclo **while**, hasta la siguiente instrucción del programa.

Veamos como sería esto en código Python:

```python
temperatura_cafe = 47 #temperatura en Celsius  

# Agregue un ciclo while que:
#  1. Muestre la temperatura actual del café, 
#  2. Muestre el mensaje "agregando hielo"; y 
#  3. Reste 1 a temperatura_cafe con cada iteración. 
# El ciclo debe ejecutarse mientras temperatura_cafe sea mayor a 43, en otras palabras, 
# debe detenerse cuando temperatura_cafe sea menor o igual a 43.

print("El café ya se puede tomar.")
```

#### Contadores

En un ciclo **while** se usan con frecuencia variables que funcionan como contadores para indicar el fin de las repeticiones. Vean el siguiente ejemplo:

*¿Cuántas veces se repite?*

```python
i = 4 #iniciar el contador
while i < 9: #definir el límite del contador
    print(i) #instrucciones que se repiten
    i = i+1 #se incrementa el contador
```

### Señalar el fin de repetición

Es común que dentro de un programa de Python se le solicite al usuario ingresar un dato que indique el fin de una repetición. Vea el siguiente ejemplo en código:

```python
suma = 0 #iniciar la sumatoria

n = input("Ingrese un número o fin para terminar: ") #pedir dato al usuario

while n != "fin":
    n_numerico = int(n) #convertir el dato a un número entero
    suma = suma + n_numerico
    n = input("Ingrese un número o fin para terminar: ")

print("Sumatoria: ",suma)
```

El siguiente es ejemplo de un ciclo infinito. Es decir un ciclo cuya expresión booleana jamás tiene un valor falso. ¿Cómo podría modificarse el siguiente programa para que no se quede infinitamente pidiendo números? 

```python
var = 1
while var == 1:
   numero = input("Ingrese un número: ")
   print ("Número ingresado: ", numero)

print ("FIN")
```

Una aplicación útil e interesante de código con repeticiones es el uso de ciclos anidados. Consideremos un gavetero:


El siguiente ciclo permitiría "revisar" el contenido de cada gaveta en el gavetero.

```python
gaveta = 1
while gaveta <= 3:
    print("Revisando la gaveta " + str(gaveta))
    gaveta += 1
```

Si tenemos tres gaveteros, cada uno con tres gavetas, ¿cómo podemos automatizar el recorrido (es decir, la revisión) de todas las gavetas? Con ciclos anidados.



```python
gavetero = 1
while gavetero <= 3:
    gaveta = 1
    while gaveta <= 3:
        print("Revisando la gaveta " + str(gaveta) + " del gavetero " + str(gavetero))
        gaveta += 1
    gavetero += 1
```
